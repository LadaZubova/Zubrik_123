```python
import os
# забираем данные картинок путь указываем наш
images = [os.path.join('C:/noy/datasets/coco128/images', x) for x in os.listdir('C:/noy/datasets/coco128/images')]
# Так как для загрузки в нашу модель данные разметки должны быть в отдельных файлах мы дублируем их из папки с ними, для дальнейшей разбивки на тестовые, обучающие и валидационные.
labels = [os.path.join('C:/noy/datasets/coco128/labels', x) for x in os.listdir('C:/noy/datasets/coco128/labels')]
```

```python
from sklearn.model_selection import train_test_split 
# производим разделение всех наших данных на Тренировочные и валидационные
train_images, val_images, train_annotations, val_annotations = train_test_split(images, labels, test_size=0.15, random_state=42)
# валидационные данные делим на тестовые и валидационные, чтобы одни использовать для тестирования пока обучается, а вторые для тестирования после обучения
val_images, test_images, val_annotations, test_annotations = train_test_split(val_images, val_annotations, test_size=0.5, random_state=42)
```

```python
import os
# Создаем папки куда будем переносить наши разбитые на 3 набора данные фотографий
parent = 'C:/noy/new'
dirs = 'images/train', 'images/val', 'images/test', 'labels/train', 'labels/val', 'labels/test'
for dir in dirs:
    path = os.path.join(parent, dir)
    # проверяем существование папки чтобы не создавать если уже есть
    if not os.path.exists(path):
        # создаем папку
        os.makedirs(path)
```

```python

import shutil
#Утилитная функция для переноса файлов 
def move_files_to_folder(list_of_files, destination_folder):
    for f in list_of_files:
        try:
            shutil.copy(f, destination_folder)
        except:
            print(f)
            assert False

# Переносим файлы по из папкам учитывая разделение
move_files_to_folder(train_images, 'C:/noy/new/images/train')
move_files_to_folder(val_images, 'C:/noy/new/images/val')
move_files_to_folder(test_images, 'C:/noy/new/images/test')
move_files_to_folder(train_annotations, 'C:/noy/new/labels/train')
move_files_to_folder(val_annotations, 'C:/noy/new/labels/val')
move_files_to_folder(test_annotations, 'C:/noy/new/labels/test')
```

```python
from PIL import Image
from yolov5.detect import run
import os

def get_photo(link):
    try: 
        dir_name = str(run(weights='C:/noy/yolov5/runs/train/exp2/weights/best.pt', source=link, save_crop= True, project = 'C:/noy/crop_photo')) + '/crops/licence/'
        images = os.listdir(dir_name)
        return images, dir_name

    except: 
        return "no", "no"
    
def get_text(images, path, reader):
    strok=[]
    try:
        for img in images:  
            image=Image.open(path+img)
            img_text=reader.readtext(f'{path}{img}')
    except:
       image = Image.open('C:/noy/Untitled.webp') 
       img_text=[(0, 'Текст не найден', 0)]

    final_text = ""

    for _, text, __ in img_text: # проходим по всему распознанному тексту
        final_text += text + ' ' # сохраняем текст с одной картинки в одну переменную
    strok.append(final_text) # добавляем в общий список картинок
    return strok[0], image
from easyocr import Reader
reader= Reader(['en'])
img, path= get_photo('C:/noy/new/images/test/Cars0.png')
text, img= get_text(img, path, reader)
print(text)
```

```python
import PySimpleGUI as sg
from PIL import Image
from yolov5.detect import run
import os
def get_photo(link):
    try: 
        dir_name = str(run(weights='C:/noy/yolov5/runs/train/exp2/weights/best.pt', source=link, save_crop= True, project = 'C:/noy/crop_photo')) + '/crops/licence/'
        images = os.listdir(dir_name)
        return images, dir_name

    except: 
        return "no", "no"
    
def get_text(images, path, reader):
    strok=[]
    try:
        for img in images:  
            image=Image.open(path+img)
            img_text=reader.readtext(f'{path}{img}')
    except:
       image = Image.open('C:/noy/Untitled.webp') 
       img_text=[(0, 'Текст не найден', 0)]

    final_text = ""

    for _, text, __ in img_text: # проходим по всему распознанному тексту
        final_text += text + ' ' # сохраняем текст с одной картинки в одну переменную
    strok.append(final_text) # добавляем в общий список картинок
    return strok[0], image
from easyocr import Reader
reader= Reader(['en'])

filename = sg.popup_get_file('Укажи файл, в по которому будем определять автомобильный номер')
f = open('C:/noy/1.txt','w')
f.write(filename)
#f.read()
img, path= get_photo(filename)
text, img= get_text(img, path, reader)
layout = [  [sg.Text(text)],     
            [sg.Button('Закрыть')] ]
window = sg.Window('Window Title', layout)  
event, values = window.read()                  
window.close()  
```
