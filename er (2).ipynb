```python
import shutil
#Утилитная функция для переноса файлов 
def move_files_to_folder(list_of_files, destination_folder):
    for f in list_of_files:
        try:
            shutil.copy(f, destination_folder)
        except:
            print(f)
            assert False

# Переносим файлы по из папкам учитывая разделение
move_files_to_folder(train_images, 'C:/123456/new/images/train')
move_files_to_folder(val_images, 'C:/123456/new/images/val')
move_files_to_folder(test_images, 'C:/123456/new/images/test')
move_files_to_folder(train_annotations, 'C:/123456/new/labels/train')
move_files_to_folder(val_annotations, 'C:/123456/new/labels/val')
move_files_to_folder(test_annotations, 'C:/123456/new/labels/test')
```

```python
from PIL import Image
from yolov5.detect import run
import os

def get_photo(link):
    try: 
        dir_name = str(run(weights='C:/123456/yolov5/runs/train/exp2/weights/best.pt', source=link, save_crop= True, project = 'C:/123456/crop_photo')) + '/crops/licence/'
        images = os.listdir(dir_name)
        return images, dir_name

    except: 
        return "no", "no"
    
def get_text(images, path, reader):
    strok=[]
    try:
        for img in images:  
            image=Image.open(path+img)
            img_text=reader.readtext(f'{path}{img}')
    except:
       image = Image.open('C:/123456/Untitled.webp') 
       img_text=[(0, 'Текст не найден', 0)]

    final_text = ""

    for _, text, __ in img_text: # проходим по всему распознанному тексту
        final_text += text + ' ' # сохраняем текст с одной картинки в одну переменную
    strok.append(final_text) # добавляем в общий список картинок
    return strok[0], image
from easyocr import Reader
reader= Reader(['en'])
img, path= get_photo("C:/123456/new/images/test/Cars0.png")
text, img= get_text(img, path, reader)
print(text)
```

```python
from yolov5.detect import run 
import os
from easyocr import Reader
# функция для получения результата
def get_number(path):
    dir_name = str(run(weights='C:/12/yolov5/runs/train/exp/weights/best.pt', source=path,save_crop= True, project = 'C:/12/crop_photo')) + '/crops/licence/'
    reader = Reader(['en','ru'], gpu = True)
    imgs = os.listdir(dir_name)
    strok = []

    for img in imgs:
        print(img)
        img_text = reader.readtext(f'{dir_name}/{img}')
        final_text = ""

        for _, text, __ in img_text: # _ = bounding box, text = text and __ = confident level
            final_text += " "
            final_text += text
        strok.append(final_text)
    return strok
print(text)
```
